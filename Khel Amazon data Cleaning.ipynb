## create books Category

```python
def assign_category(title):
    """
    Rule-based book categorizer.
    Order matters — first matching rule wins.
    """
    t = str(title).lower()

    # Rule 1: Degree / Semester Guides (B.P.Ed, M.P.Ed etc.)
    if re.search(r'\bb\.p\.ed\b|b\.p\.es|m\.p\.ed|b\.p\.e\b|semester|bped guide|mped guide|bpes guide', t):
        return 'PE Degree & Semester Guides'

    # Rule 2: Rules of Games & Sports
    if re.search(r'khelo ke niyum|khelo ke niyam|rules of games|rules of sports|niyum|niyam|sports laws|khelkud ke|laws of cricket', t):
        return 'Rules of Games & Sports'

    # Rule 3: Anatomy & Physiology
    if re.search(r'sharir rachna|anatomy|physiology|sharir kriya|exercise physiology|human anatomy', t):
        return 'Anatomy & Exercise Physiology'

    # Rule 4: Sports Psychology & Sociology
    if re.search(r'manovigyan|sports psychology|sports sociology|samajshastra|psychology in physical|khel manovigyan', t):
        return 'Sports Psychology & Sociology'

    # Rule 5: Health, Yoga & Wellness
    if re.search(r'swasthya|health education|yoga|yog |yogic|wellness|fitness', t):
        return 'Health, Yoga & Wellness'

    # Rule 6: Sports Training, Coaching & Specific Sports
    if re.search(r'prashikshan|sports training|khel chikitsa|sports medicine|sports injuries|coaching|officiating|athletics|karate|hockey|basketball|kabaddi|kho kho|cricket|chess|table tennis|boxing|tournament', t):
        return 'Sports Training, Coaching & Specific Sports'

    # Rule 7: PE Competitive Exam Books (biggest segment)
    if re.search(r'ugc|tgt|pgt|kvs|nvs|dsssb|htet|competitive exam|sampurn sharirik|sharirik shiksha ek samagra|physical education universe|physical education a key|fundamentals of physical|essentials of physical|physical education and sports science|physical education for all|physical education concepts|master stroke|champion', t):
        return 'PE Competitive Exam Books (UGC/TGT/PGT)'

    # Rule 8: General Physical Education
    if re.search(r'physical education|sharirik shiksha|sharirik shiksa|khel|sports', t):
        return 'General Physical Education'

    # Rule 9: Library Science
    if re.search(r'library|information science', t):
        return 'Library & Information Science'

    # Rule 10: General Competitive / Academic
    if re.search(r'general knowledge|samanya gyan|upsc|haryana|ias|ssc|sociology|political science|geography|bhugol|economics|arthashastra|history|itihas|home science|grah vigyan|rajniti|hindi|mathematics|biology|chemistry|physics|commerce|mba|bba|computer|digital logic', t):
        return 'General Competitive & Academic Books'

    # Rule 11: Education & Pedagogy
    if re.search(r'pedagogy|shiksha|teacher education|b\.ed|m\.ed|bal vikas|child development|educational|curriculum', t):
        return 'Education & Pedagogy'

    return 'Other / Miscellaneous'


# Apply to the Book Title column
df['Book Category'] = df['book_title'].apply(assign_category)

print('✅ Book Category column added!')
print()
print('Category breakdown:')
print(df['Book Category'].value_counts())
```

## New columns: Year / Quarter / Month

```python
# Convert date column to proper datetime
# dayfirst=True handles DD-MM-YYYY format
df['purchase_date'] = pd.to_datetime(df['purchase_date'], dayfirst=True, errors='coerce')

# Extract Year, Quarter, Month
df['Year']    = df['purchase_date'].dt.year.astype('Int64')
df['Quarter'] = df['purchase_date'].dt.quarter.apply(lambda q: f'Q{int(q)}')
df['Month']   = df['purchase_date'].dt.strftime('%B')       # e.g. 'June'
df['Month_Num'] = df['purchase_date'].dt.month              # for sorting

print('✅ Year / Quarter / Month columns added!')
print()
print('Year range:', df['Year'].min(), '→', df['Year'].max())
print('Quarters:',  df['Quarter'].unique())
df[['purchase_date','Year','Quarter','Month']].head()
```

```python
from sqlalchemy import create_engine, text
import psycopg2
import re

# ── CONNECTION CONFIG ────────────────────────────────────────────────────
username   = "postgres"
password   = "Password"       # ← your password
host       = "localhost"
port       = "5432"
database   = "khelbooks"      # lowercase — PostgreSQL best practice

# ── STEP 1: Clean column names before loading ────────────────────────────
def clean_col(name):
    name = name.strip()                    # remove trailing spaces
    name = name.lower()                    # lowercase
    name = re.sub(r'\s+', '_', name)       # spaces → underscore
    name = re.sub(r'[^a-z0-9_]', '', name) # remove special chars
    return name

df_pg = df.copy()
df_pg.columns = [clean_col(c) for c in df_pg.columns]

print("✅ Cleaned column names:")
print(df_pg.columns.tolist())

# ── STEP 2: Create database if it doesn't exist ──────────────────────────
try:
    default_engine = create_engine(
        f"postgresql+psycopg2://{username}:{password}@{host}:{port}/postgres"
    )
    with default_engine.connect() as conn:
        conn.execute(text("commit"))       # exit any open transaction
        conn.execute(text(f"CREATE DATABASE {database}"))
        print(f"✅ Database '{database}' created.")
except Exception as e:
    if "already exists" in str(e):
        print(f"ℹ️  Database '{database}' already exists — skipping creation.")
    else:
        raise e
finally:
    default_engine.dispose()

# ── STEP 3: Connect to your database ────────────────────────────────────
engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}",
    echo=False
)

# ── STEP 4: Push DataFrame to PostgreSQL ────────────────────────────────
table_name = "amazon_sales"     # lowercase, underscore — SQL standard

df_pg.to_sql(
    table_name,
    engine,
    if_exists="replace",        # replace table if it exists
    index=False,                # don't write the pandas row index
    chunksize=500,              # load in batches (safer for large files)
    method="multi"              # faster bulk insert
)

print(f"✅ {len(df_pg):,} rows loaded into table '{table_name}' in database '{database}'")
print()

# ── STEP 5: Verify — quick row count check ──────────────────────────────
with engine.connect() as conn:
    result = conn.execute(text(f"SELECT COUNT(*) FROM {table_name}"))
    count = result.scalar()
    print(f"✅ PostgreSQL confirms: {count:,} rows in '{table_name}'")

engine.dispose()
```
